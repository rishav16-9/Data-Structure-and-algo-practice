```python
# string validation
def validate(s):
    if (len(s)<10):
        return 0
    sp = "@#$-*."
    count1 = 0
    count2 = 0
    count3 = 0
    count4 = 0
    for i in s: 
        if i != i.lower():
            count1 += 1
        else:
            if i.isnumeric():
                count2 += 1
            elif i in sp:
                count3 += 1
            else:
                count4 += 1
    if (count1 !=0 and count2 !=0 and count3 !=0 and count4 !=0):
        return 1
    else:
        return 0
s=input("enter")
print(validate(s))
```

```python
#count distinctitem

# Input: l = [10,10,10]
# output = 1

# Input: l = [10,20,30]
# output = 3

def countDistinct(l):
    count = 1
    for i in range(1, len(l)):
        if(l[i] not in l[0:i]):
            count = count + 1
    return count
l = [10,20,10,30,30,20]
print(countDistinct(l))
```

 # Sub array with 0 sum in python


# Method 1

```python
# Input: l = [1,4,13,-3,-10,5]
# Output: True     13,-3,-10 = True

# Input: l = [3,1,-2,5,6]
# Output: False
    
def isSumZero(l):
    n = len(l)
    for i in range(n): 
        for j in range(i+1, n+1): 
            if(sum(l[i:j]) == 0):
                return True
    return False

# l = [3,1,-2,5,6]
# l = [1,4,13,-3,-10,5]
l = [1,4,13,-3,-10,10]

isSumZero(l)

# Time complexity is o(n^2)
```

# Method 2 (Linear time) using hashing

```python
def isSumZero(list):
    pre_sum = 0
    h = set()
    for i in range(0, len(list)):
        pre_sum += list[i]
        if(pre_sum == 0 or pre_sum in h):
            return True
        h.add(pre_sum)
    return False

# l = [1,4,13,-3,-10,10]
l = [3, -2, -1]

isSumZero(l)

# Time complexity is o(n)
```

# Check for palindorme permutation

# I/P : s= "geeks"
# O/P : False  cannot be arranged in any for 
    
# I/P: s = "geegg"
# O/P: True  gegeg or eggge

# I/P: s = "abccbabb"
# O/P: True abbccbba or bbaccabb
    
# if there in only one odd frequecy present in string then only it can be palindrome permutation

```python
from collections import Counter

def isPalPre(s):
    #line 5 and 6 Time complexity is theta(n)
    cnt = Counter(s)
    odd = 0
    #line 8 Time complexity is 0(n)
    for freq in cnt.values():
        if (freq % 2 != 0):
            odd += 1
            if (odd > 1):
                return False
    return True

# s = "abccbabb"
s = "aaabbbccaa"
isPalPre(s)

# Finak time complexity is theta(n)
```

# Check if String is ISOGRAM - It means no letter occur more than once

# Input: s = "machine"     
# output: True                   
# Input: s = "letter"
# Output: False

```python
from collections import Counter

def isIsogram(s):
    cnt = Counter(s)
    for freq in cnt.values():
        if(freq != 1):
            return False
    return True

# s ="machine"
s = "letter"
isIsogram(s)
```

# Factorial of n using recursion
# Input: 5
# Output: 120 (5* 4* 3* 2* 1= 120)

```python
def factorial(n):
    if(n == 0 or n == 1):
        return n
    
    return n * factorial(n-1)
n = int(input())
factorial(n)
```

# Find nCr 
# You are given two numbers n and r. You need to find nCr.
# nCr = (n!) / ((n-r)!*(r!))
# In recursive way, we can write nCr as nCr = (n-1)C(r-1) + (n-1)Cr

# Example 1:

# Input:
# n = 5, r = 2
# Output: 10
# Example 2:

# Input:
# n = 4, r = 1
# Output: 4

```python
def ncr(n, r):
    if(r > n):
        return 0
    elif(r == 0 and r == n):
        return 1
    elif(r == 1):
        return n
    return ncr(n-1, r-1) + ncr(n - 1, r)
ncr(5,2)
```

# Tower Of Hanoi 

# The tower of Hanoi is a famous puzzle where we have three rods and N disks. The objective of the puzzle is to move the entire stack to another rod. You are given the number of discs N. Initially, these discs are in the rod 1. You need to print all the steps of discs movement so that all the discs reach the 3rd rod. Also, you need to find the total moves.
# Note: The discs are arranged such that the top disc is numbered 1 and the bottom-most disc is numbered N. Also, all the discs have different sizes and a bigger disc cannot be put on the top of a smaller disc. Refer the provided link to get a better clarity about the puzzle.

# Example 1:

# Input:
# N = 2
# Output:
# move disk 1 from rod 1 to rod 2
# move disk 2 from rod 1 to rod 3
# move disk 1 from rod 2 to rod 3
# 3
# Explanation: For N=2 , steps will be
# as follows in the example and total
# 3 steps will be taken.
# Example 2:

# Input:
# N = 3
# Output:
# move disk 1 from rod 1 to rod 3
# move disk 2 from rod 1 to rod 2
# move disk 1 from rod 3 to rod 2
# move disk 3 from rod 1 to rod 3
# move disk 1 from rod 2 to rod 1
# move disk 2 from rod 2 to rod 3
# move disk 1 from rod 1 to rod 3
# 7
# Explanation: For N=3 , steps will be
# as follows in the example and total
# 7 steps will be taken.

```python
def toh(n, fromm, aux, to):
    if(n == 1):
        print("move disk", n, "from rod", fromm, "to rod", to)
        return 1
    count = 0
    count += toh(n-1, fromm, aux, to)
    print("move disk", n, "from rod", fromm, "to rod", to)
    count += 1
    count += toh(n-1, aux, to, fromm)
    return count
n = int(input())
print(toh(n, 1, 3, 2))
```

Josephus problem 

Given the total number of persons n and a number k which indicates that k-1 persons are skipped and kth person is killed in circle in a fixed direction.

The task is to choose the safe place in the circle so that when you perform these operations starting from 1st place in the circle, you are the last one remaining and survive.

Example 1:

Input:
n = 3 k = 2
Output: 3
Explanation: There are 3 persons so 
skipping 1 person i.e 1st person 2nd 
person will be killed. Thus the safe 
position is 3.
 

Example 2:

Input:
n = 5 k = 3
Output: 4
Explanation: There are 5 persons so 
skipping 2 person i.e 3rd person will 
be killed. Thus the safe position is 4

To solve above we use extented joseph problem

```python
def josephus(n, k):
    if(n == 1):
        return 1
    else:
        return(josephus(n-1, k) + k -1) % n + 1
n = int(input())
k = int(input())
josephus(n, k)
```

Simple josephus problem


```python
def josephus(n, k):
    if(n == 1):
        return 0
    else:
        return(josephus(n-1, k) + k) % n
n = int(input())
k = int(input())
josephus(n, k)
```
