```python
#count distinctitem

# Input: l = [10,10,10]
# output = 1

# Input: l = [10,20,30]
# output = 3

def countDistinct(l):
    count = 1
    for i in range(1, len(l)):
        if(l[i] not in l[0:i]):
            count = count + 1
    return count
l = [10,20,10,30,30,20]
print(countDistinct(l))
```

 # Sub array with 0 sum in python


# Method 1

```python
# Input: l = [1,4,13,-3,-10,5]
# Output: True     13,-3,-10 = True

# Input: l = [3,1,-2,5,6]
# Output: False
    
def isSumZero(l):
    n = len(l)
    for i in range(n): 
        for j in range(i+1, n+1): 
            if(sum(l[i:j]) == 0):
                return True
    return False

# l = [3,1,-2,5,6]
# l = [1,4,13,-3,-10,5]
l = [1,4,13,-3,-10,10]

isSumZero(l)

# Time complexity is o(n^2)
```

# Method 2 (Linear time) using hashing

```python
def isSumZero(list):
    pre_sum = 0
    h = set()
    for i in range(0, len(list)):
        pre_sum += list[i]
        if(pre_sum == 0 or pre_sum in h):
            return True
        h.add(pre_sum)
    return False

# l = [1,4,13,-3,-10,10]
l = [3, -2, -1]

isSumZero(l)

# Time complexity is o(n)
```

# Check for palindorme permutation

# I/P : s= "geeks"
# O/P : False  cannot be arranged in any for 
    
# I/P: s = "geegg"
# O/P: True  gegeg or eggge

# I/P: s = "abccbabb"
# O/P: True abbccbba or bbaccabb
    
# if there in only one odd frequecy present in string then only it can be palindrome permutation

```python
from collections import Counter

def isPalPre(s):
    #line 5 and 6 Time complexity is theta(n)
    cnt = Counter(s)
    odd = 0
    #line 8 Time complexity is 0(n)
    for freq in cnt.values():
        if (freq % 2 != 0):
            odd += 1
            if (odd > 1):
                return False
    return True

# s = "abccbabb"
s = "aaabbbccaa"
isPalPre(s)

# Finak time complexity is theta(n)
```

# Check if String is ISOGRAM - It means no letter occur more than once

# Input: s = "machine"     
# output: True                   
# Input: s = "letter"
# Output: False

```python
from collections import Counter

def isIsogram(s):
    cnt = Counter(s)
    for freq in cnt.values():
        if(freq != 1):
            return False
    return True

# s ="machine"
s = "letter"
isIsogram(s)
```

# Factorial of n using recursion
# Input: 5
# Output: 120 (5* 4* 3* 2* 1= 120)

```python
def factorial(n):
    if(n == 0 or n == 1):
        return n
    
    return n * factorial(n-1)
n = int(input())
factorial(n)
```
